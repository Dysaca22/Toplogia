```python
import numpy as np
import math
import matplotlib.pyplot as plt

matriz = np.array(
    [
      [9.55602248, 0.27949724],  
      [9.22307982, 0.43241954], 
      [8.89761026, 0.03238148],
      [3.81696476, 0.08783591],
     
      
    ]
)
```

```python
def hallar_distancia(punto_1, punto_2):
  distancia = math.sqrt( ( punto_2[0] - punto_1[0] )**2 + ( punto_2[1] - punto_1[1] )**2 )
  return distancia
```

```python
delta = 2.5
```

```python
linea = []

for i in range(0, len(matriz)):
  conexiones_de_i = []
  for j in range(0, len(matriz)):
    if i != j:
      distancia = hallar_distancia(matriz[i], matriz[j])
      if distancia <= delta:
        conexiones_de_i.append(matriz[j])
  linea.append(conexiones_de_i)
```
